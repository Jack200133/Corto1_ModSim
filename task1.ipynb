### 1. ¿Es este un escenario que se pueda modelar como una variable aleatoria de Poisson? ¿Por qué?
SI se puede poder modelar como una variable aleatoria de Poisson, la razon es que los eventos ( osea los huracanes ) ocurren de manera independiente en el tiempo, y el numero promedio de eventos es conocido. Como estamos buscando el numero de eventos que ocurren en un invtervalo de tiempo fijo ( 1 año ) y el numero promedio de eventos es conocido, entonces podemos usar la distribucion de Poisson para modelar este escenario.

### 2. Considere que usted analizará hasta un máximo de 16 huracanes este año. Grafique PMF (probability mass function) de estos eventos

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# lambda para la distribución de Poisson
lmbda = 7
x = np.arange(0, 17)

# PMF (probability mass function)
pmf_values = poisson.pmf(x, lmbda)

plt.figure(figsize=(12,6))
plt.subplot(1, 2, 1)
plt.stem(x, pmf_values, use_line_collection=True)
plt.title('PMF de la distribución de Poisson')
plt.xlabel('Número de huracanes')
plt.ylabel('Probabilidad')
```

### 3. Considere que usted analizará hasta un máximo de 16 huracanes este año. Grafique CDF (cumulative distribution function) de estos eventos

```python
# CDF (cumulative distribution function)
cdf_values = poisson.cdf(x, lmbda)

plt.subplot(1, 2, 2)
plt.step(x, cdf_values)
plt.title('CDF de la distribución de Poisson')
plt.xlabel('Número de huracanes')
plt.ylabel('Probabilidad acumulada')

plt.tight_layout()
plt.show()
```

### 4. ¿Qué conclusiones puede sacar al observar las gráficas de los ejercicios anteriores

Al observar las gráficas de la PMF y la CDF, se puede concluir que:

* PMF: La probabilidad más alta es para alrededor de 7 huracanes en un año, que es el promedio histórico. La probabilidad disminuye a medida que el número de huracanes se aleja de este promedio.

* CDF: La probabilidad acumulada de que ocurran hasta 7 huracanes es de aproximadamente 0.6. Esto significa que hay un 60% de probabilidad de que ocurran 7 huracanes o menos en un año. La probabilidad acumulada aumenta a medida que consideramos más huracanes, acercándose a 1 cuando consideramos hasta 16 huracanes.

En general, estas gráficas nos ayudan a comprender las probabilidades de diferentes números de huracanes que podrían ocurrir en un año, según la distribución de Poisson.
